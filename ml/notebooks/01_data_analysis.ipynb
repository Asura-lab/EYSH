```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pymongo import MongoClient

# Style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('husl')
```

```python
# MongoDB холболт
client = MongoClient('mongodb://localhost:27017')
db = client['eysh']

print('Collections:', db.list_collection_names())
```

```python
# Тестийн түүх татах
test_sessions = list(db.test_sessions.find())
print(f'Total test sessions: {len(test_sessions)}')

if test_sessions:
    df_sessions = pd.DataFrame(test_sessions)
    display(df_sessions.head())
```

```python
# Асуултын банк татах
questions = list(db.questions.find())
print(f'Total questions: {len(questions)}')

if questions:
    df_questions = pd.DataFrame(questions)
    
    # Сэдвээр бүлэглэх
    print('\nQuestions by subject:')
    print(df_questions['subject'].value_counts())
    
    # Хүндийн зэргээр
    print('\nQuestions by difficulty:')
    print(df_questions['difficulty'].value_counts().sort_index())
```

```python
# Visualization: Difficulty distribution
if questions:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Difficulty distribution
    axes[0].hist(df_questions['difficulty'], bins=10, edgecolor='black')
    axes[0].set_xlabel('Difficulty')
    axes[0].set_ylabel('Count')
    axes[0].set_title('Question Difficulty Distribution')
    
    # Subject distribution
    df_questions['subject'].value_counts().plot(kind='bar', ax=axes[1])
    axes[1].set_xlabel('Subject')
    axes[1].set_ylabel('Count')
    axes[1].set_title('Questions by Subject')
    axes[1].tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()
```
